# Basic

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
```

### save

```python
fig.savefig('test.png')
```

### subplot

* gcf: get current figure
* gca: get current axes
* fig: figure
* ax: axes

```python
plt.figure()
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.show()
```

```python
plt.figure()
plt.subplot(1, 2, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))
plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.show()
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
plt.show()
```

# Concat

```python
import numpy as np
import pandas as pd
```

```python
def make_df(cols, ind):
    data = {c: [str(c) + str(i) for i in ind] for c in cols}
    return pd.DataFrame(data, ind)
```

```python
make_df('ABC', range(3))
```

```python
class display(object):
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_()) for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a)) for a in self.args)
```

```python
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]
print(np.concatenate([x, y, z]))
```

```python
x = [[1, 2], [3, 4]]
print(np.concatenate([x, x]))
```

```python
x = [[1, 2], [3, 4]]
print(np.concatenate([x, x], axis=0))
```

```python
x = [[1, 2], [3, 4]]
print(np.concatenate([x, x], axis=1))
```

```python
ser1 = pd.Series(['A', 'B', 'C'], index=[1, 2, 3])
ser2 = pd.Series(['D', 'E', 'F'], index=[4, 5, 6])
print(pd.concat([ser1, ser2]))
```

```python
df1 = make_df('AB', [1, 2])
df2 = make_df('AB', [3, 4])
display('df1', 'df2', 'pd.concat([df1, df2])')
```

```python
df3 = make_df('AB', [0, 1])
df4 = make_df('CD', [0, 1])
display('df3', 'df4', "pd.concat([df3, df4], axis=1)")
```

```python
x = make_df('AB', [0, 1])
y = make_df('AB', [2, 3])
y.index = x.index  # make duplicate indices
display('x', 'y', 'pd.concat([x, y])')
```

```python
try:
    print(pd.concat([x, y], verify_integrity=True))
except ValueError as e:
    print(e)
```

```python
display('x', 'y', 'pd.concat([x, y], ignore_index=True)')
```

```python
display('x', 'y', "pd.concat([x, y], keys=['x', 'y'])")
```

### join
* outer: an union of the input columns (default)
* inner: an intersection of the columns

```python
df5 = make_df('ABC', [1, 2])
df6 = make_df('BCD', [3, 4])
display('df5', 'df6', 'pd.concat([df5, df6])')
```

```python
display('df5', 'df6', "pd.concat([df5, df6], join='inner')")
```

```python
display('df5', 'df6', "pd.concat([df5, df6], join_axes=[df5.columns])")
```
